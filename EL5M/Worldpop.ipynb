# WorldPop Global Project Population Data: 
## Constrained Estimated Age and Sex Structures of Residential Population 
## per 100x100m Grid Square 
[WorldPop/GP/100m/pop_age_sex_cons_unadj](https://developers.google.com/earth-engine/datasets/catalog/WorldPop_GP_100m_pop_age_sex_cons_unadj?hl=en)

Currently only 2020 data are present.

```python
import os
import ee
ee.Initialize()
import leafmap
#leafmap.update_package()
import geemap
```

```python
Map = geemap.Map(center = (40, -100), zoom = 3)
Map
```

```python
Map.add_basemap('SATELLITE')
```

```python
gaul0 = ee.FeatureCollection("FAO/GAUL/2015/level0")
gaul1 = ee.FeatureCollection("FAO/GAUL/2015/level1")
gaul2 = ee.FeatureCollection("FAO/GAUL_SIMPLIFIED_500m/2015/level2")
roi = gaul1.filter(ee.Filter.eq('ADM0_NAME', 'Mexico'))
Map.centerObject(roi)
```

```python
merit = ee.Image("MERIT/DEM/v1_0_3").select('dem')
lowAreas = merit.lte(5)
Map.addLayer(lowAreas.selfMask(), {'min': 0, 'max':1, 'palette': ['grey','blue']}, "Low Areas")
```

```python
worldpop = ee.ImageCollection("WorldPop/GP/100m/pop_age_sex_cons_unadj").select('population')
visualization = {
  'bands': ['population'],
  'min': 0.0,
  'max': 50.0,
  'palette': ['24126c', '1fff4f', 'd4ff50']
};
Map.addLayer(worldpop, visualization, "Population");
```

```python
worldpop_lowAreas=worldpop.mean().mask(lowAreas)

visualization = {
  'bands': ['population'],
  'min': 0.0,
  'max': 50.0,
  'palette': ['24126c', '1fff4f', 'd4ff50']
};
Map.addLayer(worldpop_lowAreas, visualization, "Population Low Areas");
```

```python
out_dir = os.path.join(os.path.expanduser('~'), 'Downloads')
out_dem_stats = os.path.join(out_dir, 'worldpop.csv')  

if not os.path.exists(out_dir):
    os.makedirs(out_dir)

# Allowed output formats: csv, shp, json, kml, kmz
# Allowed statistics type: MEAN, MAXIMUM, MINIMUM, MEDIAN, STD, MIN_MAX, VARIANCE, SUM
geemap.zonal_statistics(worldpop, roi, out_dem_stats, statistics_type='SUM', scale=100)
geemap.create_download_link(out_dem_stats)
```

```python
out_dir = os.path.join(os.path.expanduser('~'), 'Downloads')
out_dem_stats = os.path.join(out_dir, 'worldpop_lowAreas.csv')  

if not os.path.exists(out_dir):
    os.makedirs(out_dir)

# Allowed output formats: csv, shp, json, kml, kmz
# Allowed statistics type: MEAN, MAXIMUM, MINIMUM, MEDIAN, STD, MIN_MAX, VARIANCE, SUM
geemap.zonal_statistics(worldpop_lowAreas, roi, out_dem_stats, statistics_type='SUM', scale=100)
geemap.create_download_link(out_dem_stats)
```
